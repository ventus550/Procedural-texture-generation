### Generate chaotic shapes and melted edges
1. Test diffrent applications of wolrey noise
2. Run some experiments

```python
import __init__
import numpy as np
from engine import Heatmap, Noise
from noises import smooth, heat, worley
from utilities import *
```

```python
plates = worley(500, 100, 2)
landmass = heat(500, 20, 250, 250)

Heatmap(plates, landmass, 2*plates + landmass + 0.5*smooth(10, 50), scale = 2.0, cmap = 'frostfire')
```

```python
def rifts(orbs, clusters):
	return 2*worley(400, orbs,1) - worley(400, clusters, 2)
```

```python
Noise(rifts, orbs=(2, 300, 5), clusters=(2, 100), cmap = 'frostfire')
```

```python
rfts = rifts(80, 5)
Heatmap(rfts, np.cos(rfts * 2*np.pi), np.sin(rfts * 2*np.pi), scale = 2.0, cmap = 'frostfire')
```

```python
volcano = -heat(500, 20, 250, 500)
sm = smooth(5, 100)
rivers = vectorize(sm, alpha = 5, time=80) + 1

Heatmap(sm, 1/rivers, sm + norm(1/rivers)/2 + volcano, scale = 2.0, cmap = 'frostfire')
```

```python
plates = worley(500, 50, 1)
volcano = -heat(500, 21, 250, 250)
sm = smooth(4, 125)
rivers = vectorize(volcano, alpha = 5, time=80) + 1
Heatmap(sm, 1/rivers, softmax(volcano + norm(1/rivers)) + sm / 100000, scale = 1.5, cmap = 'frostfire')
```
