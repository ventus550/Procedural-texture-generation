### Generate more rocky looking mountains
1. Create roundy shapes with wolrey noise
2. Combine them with various simplex noises

```python
import __init__
import numpy as np
from engine import Heatmap, Noise, Field, gradient
from noises import smooth, heat, worley, simplex
from utilities import *
```

```python
plates = 1 - worley(500, 100, 1)
landmass = heat(500, 20, 250, 250) 

Heatmap(plates, landmass, 4*plates + landmass, scale = 2.0, cmap='snowpeaks')
```

```python
plates = 1 - worley(500, 100, 1)
landmass = simplex(500, octaves=9)

Heatmap(plates, landmass, 2*plates + landmass, scale = 2.0, cmap='snowpeaks')
```

```python
Noise(simplex,
	resolution = (1, 1000),
	octaves = 8,
	persistence = 5.0,
	lacunarity = 0.5,
	seed = (-1000, 1000), cmap = 'snowpeaks')
```

```python
plates = 1 - worley(500, 100, 1)
Heatmap(*[simplex(500, oct, 3.20, 0.6) + plates for oct in (9, 10, 11, 12)], scale = 2.0, cmap = 'snowpeaks')
Heatmap(*[simplex(500, oct, 3.20, 0.6) + plates for oct in (9, 10, 11, 12)], scale = 2.0, cmap = 'snowpeaks')
Heatmap(*[simplex(500, oct, 3.20, 0.6) + plates for oct in (9, 10, 11, 12)], scale = 2.0, cmap = 'snowpeaks')
```

```python
simp = simplex(500, 12, 3.20, 0.6)
Heatmap(*[simp + (1 - worley(500, orbs, 1)) for orbs in (2, 4, 16, 32, 64, 128)], scale = 2.0, cmap = 'snowpeaks')
```

```python
simp = simplex(500, 12, 3.20, 0.6)
Heatmap(*[simp + (1 - worley(500, 1 + n, n)) for n in (1, 2, 3, 4)], scale = 2.0, cmap = 'snowpeaks')
```
